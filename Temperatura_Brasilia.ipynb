# Gráfico de Temperatura
---
O programa a seguir visa construir um gráfico mostrando o comportamento de duas curvas distintas com o horário em relação a temperatura em determinada parte do dia.


```python
bsb_temp = [14, 16, 19, 21, 23, 25, 26, 26, 26]
nyc_temp = [18, 19, 21, 22, 23, 24, 26, 26, 26]

t = range(7, 16)
```

```python
import matplotlib.pyplot as plt

if __name__ == '__main__':
    plt.title('Temperatura-Horas')
    plt.ylabel('Graus')
    plt.xlabel('Horas')
    
    plt.plot(t, bsb_temp, marker='o')
    plt.plot(t, nyc_temp, marker='x')
    
    plt.legend(['bsb', 'nyc'])
    
    plt.show()
```
