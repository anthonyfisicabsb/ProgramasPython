# Lei da Gravitação Universal de Newton
---
A partir de seus estudos, o físico Isaac Newton descobriu que existia uma lei de caráter universal, a qual dizia que dois corpos de atraiam a partir de um fator proporcional à massa de cada um e inversamente proporcional ao quadrado das distâncias entre eles.  
F = G.M.m/r²

```python
import matplotlib.pyplot as plt
```

```python
#draw the graph
def draw_graph(x, y):
    plt.plot(x, y, marker='o')
    plt.xlabel('Distance in meters')
    plt.ylabel('Gravitacional force in Newtons')
    plt.title('Gravitacional force and distance')
    plt.show()
```

```python
def generate_F_r():
    #Generate values for R
    r = range(100, 1001, 50)
    #Empyt list to store the calculated values for F
    F = []
    
    #Constant G
    G = 6.674 * (10**-11)
    #Two masses
    m1 = 0.5
    m2 = 1.5
    
    #Calculate force and add it to the list
    for dist in r:
        force = G*(m1*m2)/(dist**2)
        F.append(force)
    
    #Call the draw_graph function
    draw_graph(r, F)
```

```python
if __name__ == '__main__':
    generate_F_r()
```
