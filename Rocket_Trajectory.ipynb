# Rocket Trajectory
---
Programa que calcula a trajetória de vários foguetes diferentes.  
O ângulo de partida considerado será de 45º, ou seja, o alcance máximo

```python
import matplotlib.pyplot as plt

def create_graph(x, y):
    plt.plot(x, y)
```

```python
from math import *

def create_list(v, theta):
    t_total = (2 * v * sin(theta))/9.78
    
    tempos = []
    t = 0.0
    while t <= t_total:
        tempos.append(t)
        t += 0.01
    
    y_list = []
    
    for tempo in tempos:
        y_list.append(v*sin(theta)*tempo - 9.78/2*(tempo**2))
    
    create_graph(tempos, y_list)
```

```python

try:
    val1 = float(input('insira a primeira velocidade em m/s: '))
    val2 = float(input('insira a segunda velocidade em m/s: '))
    val3 = float(input('insira a terceira velocidade em m/s: '))
        
    create_list(val1, 0.7854)
    create_list(val2, 0.7854)
    create_list(val3, 0.7854)
        
    plt.xlabel('tempo')
    plt.ylabel('altura')
        
    plt.title('Lançamento Oblíquo')
    plt.legend([str(val1), str(val2), str(val3)])
        
    plt.show()
except ValueError:
    print('Insira valores corretos para as velocidades')
```
