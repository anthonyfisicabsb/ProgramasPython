# Série de Fourier
---
Este programa tem a finalidade de encontrar a série de fourier para um conjunto de sinais obtidos

```python
"""Bibliotecas necessárias"""
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""Função periódica a ser analisada"""
def func(x):
    if x >= -3 and x < -1:
        return 1
    elif x >= -1 and x < 1:
        return 0
    elif x >=1 and x < 3:
        return 1
```

```python
"""Aplica a integração utilizando a regra dos trapézios"""
def integracao(lim_inf, lim_sup, fun):
    h = (lim_sup - lim_inf)/150
    resultado = (fun(lim_sup) + fun(lim_inf))/2
    
    for i in range(1, 150):
        resultado += fun(lim_inf + i*h)
    
    resultado *= h
    
    return resultado
```

```python
"""Retorna os coeficientes da série de fourier"""
def a(n, periodo):
    aux = lambda x : func(x)*np.cos((2*n*np.pi*x)/periodo)
    
    return integracao(-periodo/2.0, periodo/2.0, aux)

def b(n, periodo):
    aux = lambda x : func(x)*np.sin((2*n*np.pi*x)/periodo)
    
    return integracao(-periodo/2.0, periodo/2.0, aux)
```

```python
def fourier(periodo, x, tot=11):
    a0 = a(0, periodo)
    resultado = a0/2
    
    for i in range(1, tot):
        resultado += (a(i, periodo)*np.cos((2*i*np.pi*x)/periodo) + b(i, periodo)*np.sin((2*i*np.pi*x)/periodo))
    
    return resultado
```

```python
def create_graph(x, y, title):
    plt.title(title)
    plt.plot(x, y)
    plt.show()
```

```python
if __name__ == '__main__':
    sup = float(input('Insira o limite superior do dominio: '))
    inf = float(input('Insira o limite inferior do dominio: '))
    periodo = float(input('Insira o periodo da funcao: '))
    
    x_dom = [inf]
    
    aux = inf + 1e-3
    while aux <= sup:
        x_dom.append(aux)
        aux += 1e-3
    
    y_dom = [func(x) for x in x_dom]
    
    create_graph(x_dom, y_dom, 'Gráfico da Função')
```

```python
fourier_2 = [fourier(periodo, x, 2) for x in x_dom]

create_graph(x_dom, fourier_2, 'Serie de Fourier com n=2')
```
